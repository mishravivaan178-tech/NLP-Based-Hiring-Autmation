```python
# ============================================================
# 04_EDUCATION_EXTRACTION.IPYNB
# FINAL VERSION
# ============================================================

import pandas as pd
import re

# ============================================================
# STEP 1 : LOAD DATASET
# ============================================================

df = pd.read_csv("resumeJD_with_experience.csv")

print("="*50)
print("DATASET LOADED")
print("="*50)

print("Shape:", df.shape)

# ============================================================
# STEP 2 : EDUCATION PATTERNS
# ============================================================

EDUCATION_PATTERNS = {

    # ==========================
    # Bachelor's Degrees
    # ==========================

    "B.Tech": r"\bb\s*tech\b",
    "B.E": r"\bb\s*e\b",
    "B.Sc": r"\bb\s*sc\b",
    "B.Com": r"\bb\s*com\b",
    "B.A": r"\bb\s*a\b",
    "BBA": r"\bbba\b",
    "B.Des": r"\bb\s*des\b",

    # ==========================
    # Master's Degrees
    # ==========================

    "MBA": r"\bmba\b",
    "M.Tech": r"\bm\s*tech\b",
    "M.Sc": r"\bm\s*sc\b",
    "M.A": r"\bm\s*a\b",
    "M.S": r"\bm\s*s\b",
    "PGDM": r"\bpgdm\b",

    # ==========================
    # Professional Degrees
    # ==========================

    "CA": r"\bca\b",
    "CFA": r"\bcfa\b",

    # ==========================
    # Diploma
    # ==========================

    "Diploma": r"\bdiploma\b",

    # ==========================
    # Doctorate
    # ==========================

    "PhD": r"\bph\s*d\b"
}

# ============================================================
# STEP 3 : EXTRACT EDUCATION
# ============================================================

def extract_education(text):

    text = str(text).lower()

    for degree, pattern in EDUCATION_PATTERNS.items():

        if re.search(pattern, text):

            return degree

    return "Unknown"

# ============================================================
# STEP 4 : EDUCATION LEVEL
# ============================================================

def education_level(degree):

    bachelor = [
        "B.Tech",
        "B.E",
        "B.Sc",
        "B.Com",
        "B.A",
        "BBA",
        "B.Des"
    ]

    master = [
        "MBA",
        "M.Tech",
        "M.Sc",
        "M.A",
        "M.S",
        "PGDM"
    ]

    professional = [
        "CA",
        "CFA"
    ]

    if degree in bachelor:
        return "Bachelor"

    elif degree in master:
        return "Master"

    elif degree in professional:
        return "Professional"

    elif degree == "Diploma":
        return "Diploma"

    elif degree == "PhD":
        return "Doctorate"

    else:
        return "Unknown"

# ============================================================
# STEP 5 : CREATE NEW COLUMNS
# ============================================================

df["resume_education"] = df["resume_text"].apply(
    extract_education
)

df["resume_education_level"] = df[
    "resume_education"
].apply(
    education_level
)

print("\nEducation Extraction Completed")

# ============================================================
# STEP 6 : SAMPLE OUTPUT
# ============================================================

print("\n" + "="*50)
print("SAMPLE OUTPUT")
print("="*50)

print(

    df[
        [
            "resume_education",
            "resume_education_level"
        ]
    ].head(20)

)

# ============================================================
# STEP 7 : EDUCATION STATISTICS
# ============================================================

print("\n" + "="*50)
print("EDUCATION DISTRIBUTION")
print("="*50)

print(
    df["resume_education"]
    .value_counts()
)

print("\n" + "="*50)
print("EDUCATION LEVEL DISTRIBUTION")
print("="*50)

print(
    df["resume_education_level"]
    .value_counts()
)

# ============================================================
# STEP 8 : UNKNOWN EDUCATION CHECK
# ============================================================

unknown_count = len(

    df[
        df["resume_education"]
        ==
        "Unknown"
    ]

)

print("\nUnknown Education Count:", unknown_count)

# ============================================================
# STEP 9 : SAVE DATASET
# ============================================================

df.to_csv(
    "resumeJD_with_education.csv",
    index=False
)

print("\nDataset Saved Successfully")

print(
    "\nFile Name: resumeJD_with_education.csv"
)

# ============================================================
# STEP 10 : DOWNLOAD FILE
# ============================================================

from google.colab import files

files.download(
    "resumeJD_with_education.csv"
)
```
