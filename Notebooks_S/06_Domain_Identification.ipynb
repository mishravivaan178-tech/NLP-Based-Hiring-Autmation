```python
# ============================================================
# 06_DOMAIN_IDENTIFICATION.IPYNB
# ============================================================

import pandas as pd
import ast

# ============================================================
# STEP 1 : LOAD DATASET
# ============================================================

df = pd.read_csv(
    "resumeJD_with_semantic_similarity.csv"
)

print("="*50)
print("DATASET LOADED")
print("="*50)

print("Shape:", df.shape)

# ============================================================
# STEP 2 : DOMAIN KNOWLEDGE BASE
# ============================================================

DOMAIN_SKILLS = {

    "Finance": [
        "cfa","ca","valuation","equity research",
        "financial reporting","accounting",
        "treasury","investment banking",
        "financial analysis","financial modeling",
        "forecasting","budgeting","quickbooks",
        "sap","ifrs","gaap"
    ],

    "HR": [
        "recruitment","talent acquisition",
        "employee relations","payroll",
        "human resources","onboarding",
        "performance management",
        "training and development"
    ],

    "Marketing": [
        "seo","sem","google analytics",
        "content marketing",
        "email marketing",
        "social media marketing",
        "branding","market research"
    ],

    "Data Science": [
        "python","machine learning",
        "deep learning","tensorflow",
        "pytorch","nlp",
        "data science",
        "data analysis",
        "scikit learn"
    ],

    "Software Development": [
        "java","c++","javascript",
        "react","angular",
        "nodejs","node.js",
        "django","flask",
        "spring boot"
    ],

    "DevOps": [
        "aws","azure","gcp",
        "docker","kubernetes",
        "terraform","jenkins",
        "ansible","linux"
    ],

    "Cyber Security": [
        "cyber security",
        "penetration testing",
        "ethical hacking",
        "network security",
        "information security",
        "siem"
    ],

    "Product Management": [
        "product management",
        "product analytics",
        "roadmapping",
        "user research",
        "growth hacking"
    ],

    "Sales": [
        "sales","salesforce",
        "lead generation",
        "crm",
        "business development",
        "customer relationship management"
    ]
}

# ============================================================
# STEP 3 : FIND DOMAIN
# ============================================================

def identify_domain(skill_list):

    if isinstance(skill_list, str):

        try:
            skill_list = ast.literal_eval(
                skill_list
            )

        except:
            skill_list = []

    scores = {}

    for domain, skills in DOMAIN_SKILLS.items():

        count = len(
            set(skill_list)
            &
            set(skills)
        )

        scores[domain] = count

    best_domain = max(
        scores,
        key=scores.get
    )

    if scores[best_domain] == 0:

        return "Unknown"

    return best_domain

# ============================================================
# STEP 4 : RESUME DOMAIN
# ============================================================

df["resume_domain"] = df[
    "resume_skills"
].apply(
    identify_domain
)

# ============================================================
# STEP 5 : JD DOMAIN
# ============================================================

df["jd_domain"] = df[
    "jd_skills"
].apply(
    identify_domain
)

# ============================================================
# STEP 6 : DOMAIN MATCH
# ============================================================

df["domain_match"] = (

    df["resume_domain"]
    ==
    df["jd_domain"]

).astype(int)

print(
    "\nDomain Identification Completed"
)

# ============================================================
# STEP 7 : CHECK OUTPUT
# ============================================================

print("\nSample Output\n")

print(

    df[
        [
            "resume_domain",
            "jd_domain",
            "domain_match"
        ]
    ]
    .head(20)

)

# ============================================================
# STEP 8 : STATISTICS
# ============================================================

print("\nResume Domain Distribution\n")

print(
    df["resume_domain"]
    .value_counts()
)

print("\nJD Domain Distribution\n")

print(
    df["jd_domain"]
    .value_counts()
)

print("\nDomain Match Distribution\n")

print(
    df["domain_match"]
    .value_counts()
)

# ============================================================
# STEP 10 : SAVE & DOWNLOAD DATASET
# ============================================================

from google.colab import files

df.to_csv(
    "resumeJD_with_domains.csv",
    index=False
)

print("\nDataset Saved Successfully")

print(
    "File Name: resumeJD_with_domains.csv"
)

files.download(
    "resumeJD_with_domains.csv"
)
```
