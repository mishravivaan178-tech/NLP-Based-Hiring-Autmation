CORRELATION ANALYSIS

```python
cols=['skill_match_score','experience_match_score','semantic_similarity_score','domain_match','education_match_score']
print(df1[cols].corr()['education_match_score'])
sns.heatmap(df1[cols].corr(), annot=True, cmap='coolwarm')
print(df1['match_label'].unique())
df1['num_match_label']=df1['match_label'].replace(['partial match', 'no match', 'match'], [0.5, 0, 1])
cols2 = ['skill_match_score','experience_match_score','semantic_similarity_score','domain_match','education_match_score','num_match_label']
print(df1[cols2].corr()['num_match_label'])
sns.heatmap(df1[cols2].corr(), annot=True, cmap='coolwarm')
```

```python
dm=df1[['skill_match_score','experience_match_score','semantic_similarity_score','domain_match','education_match_score',"match_label","num_match_label"]]
dm=dm.rename(columns={'Unnamed: 0':'idx'})
print(dm.head())
print(dm.columns)
dm.to_csv("../Data/datamain.csv",index=False)
```
