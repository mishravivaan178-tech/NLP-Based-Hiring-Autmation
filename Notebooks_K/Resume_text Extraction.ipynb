```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("../Datasets/resumeJD2_pairs.csv")
if df is False:
    print("The dataset is empty.")
else:
    print("-------------------------------Dataset loaded successfully!----------------------------------")
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    print(df.columns)
    print(df.head())
print("-------------------------------Dataset Information----------------------------------")
df.info()
df.describe()
df.isnull().sum()
```

```python
print(df['resume_text'][0])
patterns = {
    "name": r"Name:\s*(.*?)\s*\|\|",
    "email": r"Email:\s*([^\s|]+)",
    "phone": r"Phone:\s*([^|]+)",
    "linkedin": r"LinkedIn:\s*([^|]+)",
    "location": r"Location:\s*([^|]+)",
    "open_to": r"Open to:\s*([^|]+)",
    "summary": r"PROFESSIONAL SUMMARY\s*\|\|\s*(.*?)\s*\|\|\s*---",
    "skills": r"SKILLS\s*\|\|\s*(.*?)\s*\|\|\s*---",
    "education": r"EDUCATION\s*\|\|\s*(.*?)(?:$|\s*\|\|\s*CERTIFICATIONS)",
}
for col, pattern in patterns.items():
    df[col]=df["resume_text"].str.extract(pattern,expand=False)
for col in patterns:
    df[col]=df[col].str.strip()

df.columns
```

```python
df.drop(['resume_text','job_description','match_score','match_label'],axis=1,inplace=True)
print(df.columns)
df.to_csv("../DataExtracted/resume_text extracted.csv",index=False)
```
